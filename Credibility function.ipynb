```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(receiver):
    
    alpha = 100 - receiver
    beta = 0
    sigma = math.exp(0.01*(receiver - beta))
    interval = alpha / (1 + sigma)
    
    return interval
```

```python
x = np.linspace(0, 100, 1000)
y = []
for i in range(len(x)):
    y.append(f(x[i]))
```

```python
plt.plot(x,y)
plt.xlabel("Education level")
plt.ylabel("Size of interval")
plt.title("\u03B1 = 0.01    \u03B2 = 0")
```
